1) Perform the entire NLP pipeline (Tokenization, Stopword Removal, Stemming, and Lemmatization) on a custom paragraph and print the results at each step.

```python
custom_paragraph = """Natural language processing is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human (natural) languages, in particular how to program computers to process and analyze large amounts of natural language data. It is a crucial part of many modern applications, such as chatbots, sentiment analysis tools, and machine translation systems. Understanding the nuances of human language is a complex task, but NLP techniques are constantly evolving to improve accuracy and efficiency."""
print(custom_paragraph)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')   # <-- add this

from nltk.tokenize import word_tokenize

tokens = word_tokenize(custom_paragraph)
print(tokens)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print(filtered_tokens)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

print(stemmed_tokens)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]

print(lemmatized_tokens)
```

2) Define NLP and its real-time applications in a specific domain.
Definition of NLP:

Natural Language Processing (NLP) is a subfield of artificial intelligence, linguistics, and computer science that focuses on enabling computers to understand, interpret, and generate human language. Its primary goal is to bridge the gap between human communication and computer comprehension, allowing machines to process, analyze, and make sense of large amounts of text and speech data.

Chosen Domain: Healthcare

Real-time NLP Applications in Healthcare:

Electronic Health Record (EHR) Analysis: Description: NLP is used to extract valuable information from unstructured clinical notes, physician dictations, and patient records. This includes identifying medical conditions, treatments, medications, and procedures. Benefits: Improves data accuracy and completeness, enables better clinical decision support, facilitates research by making patient data searchable and analyzable, and reduces the burden on healthcare professionals for manual data entry and review.

Clinical Trial Matching: Description: NLP helps identify eligible patients for clinical trials by analyzing their medical history and comparing it against trial criteria. This process can be automated to quickly and accurately find suitable candidates. Benefits: Speeds up patient recruitment for clinical trials, increases the chances of finding the right patient population, and ultimately accelerates the development of new treatments and therapies.

Medical Chatbots and Virtual Assistants: Description: NLP powers chatbots and virtual assistants that can interact with patients, answer their questions about symptoms, medications, or appointments, and provide general health information. These systems can understand patient queries in natural language and provide relevant responses. Benefits: Improves patient engagement and access to information, reduces the workload on healthcare staff, provides 24/7 support, and can help triage patient inquiries to the appropriate healthcare professional.

3) Define NLU and NLG.
Definition of Natural Language Understanding (NLU):

Natural Language Understanding (NLU) is a subfield of NLP that focuses on enabling computers to understand the meaning and intent behind human language. It involves analyzing the grammatical structure, identifying entities, and determining the relationships between words to extract meaning from text or speech.

Definition of Natural Language Generation (NLG):

Natural Language Generation (NLG) is a subfield of NLP that focuses on enabling computers to generate human-like text or speech from structured data. It involves converting data into coherent and grammatically correct sentences, paragraphs, or even longer pieces of text.
